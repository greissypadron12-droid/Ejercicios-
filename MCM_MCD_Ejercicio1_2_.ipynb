```python
# Cálculo del Mínimo Común Múltiplo (MCM)
# usando el algoritmo de Euclides para el MCD

# Datos ya definidos
numero1 = int(24)
numero2 = int(36)

# Proceso: Algoritmo de Euclides para el MCD
a = numero1
b = numero2

while b != 0:
    residuo = a % b
    a = b
    b = residuo

mcd = a  # El último valor de 'a' es el MCD

# Cálculo del MCM con la fórmula
mcm = int((numero1 * numero2) / mcd)

# Salida (resultado)
print("El Máximo Común Divisor de", numero1, "y", numero2, "es:", mcd)
print("El Mínimo Común Múltiplo de", numero1, "y", numero2, "es:", mcm)
```
