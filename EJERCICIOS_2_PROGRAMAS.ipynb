# 1.Escriba un programa que lea 3 números  y determine cual es el mayor.


**Definición de variables**

num1: primer número ingresado: 2

num2: segundo número ingresado: 8

num3: tercer número ingresado: 6

mayor: variable donde se guardará el número más grande.

**Lenguaje natural**

El programa leerá tres números: 2, 8 y 6.
Luego comparará cuál es el mayor utilizando estructuras condicionales if, elif y else.
Finalmente, mostrará en pantalla el número mayor.

Diagrama de flujo

Diagrama de flujo primer programa.png

**Código**

```python
# Programa para determinar el mayor de tres números

# Definición de variables
num1 = 2
num2 = 8
num3 = 6

# Comparación
if num1 > num2 and num1 > num3:
    mayor = num1
elif num2 > num3:
    mayor = num2
else:
    mayor = num3

# Resultado
print("Los números son:", num1, ",", num2, ",", num3)
print("El número mayor es:", mayor)
```

**Resultado:** Los números son: 2 , 8 , 6
El número mayor es: 8

# 2. Escriba un programa que determine si una palabra dada es un palíndromo.

Políndromos: Los palíndromos son palabras, frases, números o nombres que se leen igual de izquierda a derecha que de derecha a izquierda.*texto en cursiva*

**Definición de variables**

palabra → cadena de texto ingresada ("yo hago yoga hoy").

palabra_invertida → texto al revés.

resultado → mensaje que indica si es palíndromo o no.

**Lenguaje Natural**

Inicio del programa.

1. Se solicita al usuario que ingrese una palabra o frase.

2. El usuario ingresa la frase: “yo hago yoga hoy”.

3. La frase se guarda en la variable texto_original.

4. Se limpia la frase para poder analizarla correctamente:

5. Se convierten todas las letras a minúsculas.

6. Se eliminan los espacios.

7. El resultado se guarda en la variable texto_limpio, quedando así: “yohagoyogahoy”.

8. Se invierte el texto limpio y se guarda en texto_invertido, que también resulta ser “yohagoyogahoy”.

9. El programa compara si texto_limpio es igual a texto_invertido.

Como ambas cadenas son idénticas, el programa muestra el mensaje:
“La frase es un palíndromo.”

Fin del programa

Diagrama de flujo
Diagrama de flujo segundo programa.png

**Código**

```python
# Programa para determinar si una palabra o frase es un palíndromo

# Definición de variables
palabra = "yo hago yoga hoy"

# Procesamiento
palabra = palabra.replace(" ", "").lower()   # Quita espacios y convierte a minúsculas
palabra_invertida = palabra[::-1]            # Invierte la palabra

# Comparación
if palabra == palabra_invertida:
    print("La frase es un palíndromo.")
else:
    print("La frase no es un palíndromo.")
```

**Resultado: **
La frase es un palíndromo.
