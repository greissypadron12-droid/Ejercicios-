```python
# Cálculo de la velocidad
# Fórmula: velocidad = distancia / tiempo

# Datos ya definidos con tipos
distancia = int(300)   # en metros
tiempo = int(15)       # en segundos

# Proceso (cálculo de la velocidad)
velocidad = float(distancia / tiempo)

# Salida (resultado)
print("La velocidad es:", velocidad, "m/s")
```
